```python
# Load data

def load_FAE(input_path="FGNET\images", size=(300,300),show=False):
    try :
        file_names = sorted(os.listdir(input_path))#[:10]
    except : 
        input_path = input_path.replace('\\',"/")
        file_names = sorted(os.listdir(input_path))
    
    data = np.zeros((len(file_names),size[0],size[1]))
    labels = np.zeros((len(file_names)))
    #for i in tqdm(range(10)):

    for i in tqdm(range(len(file_names))) :

        file_name=file_names[i]
        labels[i] = int(file_name[4:6])
        with Image.open(os.path.join(input_path,file_name),'r') as imgpil:      
            img1 = imgpil.convert("L")
            img2 = img1.resize(size)
            if show and i<10 :
                img2.show()
            img = np.array(img2)
            data[i]=img
    #print(labels)
    return data,labels
```

```python
# Votre code ici

def modele_least_square(X,ytrain, normalize=False):
    if len(X.shape) == 2 : 
        Xreshape = X.astype('uint8')
    if len(X.shape) == 3 :
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]).astype('uint8')
    if len(X.shape) == 4 : 
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]*X.shape[3]).astype('uint8') 
    
    model = LinearRegression(normalize = normalize)
    model.fit(Xreshape, ytrain)
    return model

def modele_ridgeCV(X,ytrain, normalize=False):
    if len(X.shape) == 2 : 
        Xreshape = X.astype('uint8')
    if len(X.shape) == 3 :
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]).astype('uint8')
    if len(X.shape) == 4 : 
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]*X.shape[3]).astype('uint8') 
    
    # On test différentes valeurs de alpha
    model = RidgeCV(alphas=[1e-4,1e-3, 1e-2, 1e-1, 1, 10],normalize = normalize, store_cv_values=True)
    model.fit(Xreshape, ytrain)
    return model

def modele_lassoCV(X, ytrain, normalize=False):
    if len(X.shape) == 2 : 
        Xreshape = X.astype('uint8')
    if len(X.shape) == 3 :
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]).astype('uint8')
    if len(X.shape) == 4 : 
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]*X.shape[3]).astype('uint8') 
    
    model = LassoCV(alphas = [1e-3, 1e-2, 1e-1, 1], normalize = normalize, verbose = 1, cv = 2)
    model.fit(Xreshape, ytrain)
    return model

def modele_lasso(X, ytrain, normalize=False):
    if len(X.shape) == 2 : 
        Xreshape = X.astype('uint8')
    if len(X.shape) == 3 :
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]).astype('uint8')
    if len(X.shape) == 4 : 
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]*X.shape[3]).astype('uint8') 
    
    model = Lasso(normalize = normalize)
    model.fit(Xreshape, ytrain)
    return model

def modele_elasticnetCV(X,ytrain, normalize=False):
    if len(X.shape) == 2 : 
        Xreshape = X.astype('uint8')
    if len(X.shape) == 3 :
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]).astype('uint8')
    if len(X.shape) == 4 : 
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]*X.shape[3]).astype('uint8') 
    if normalize :
        Xreshape = norm(Xreshape)
    
    model = ElasticNetCV(alphas = [1e-3, 1e-2, 1e-1, 1], verbose = 1, cv = 2)
    model.fit(Xreshape, ytrain)
    return model

def modele_elasticnet(X,ytrain, normalize=False):
    if len(X.shape) == 2 : 
        Xreshape = X.astype('uint8')
    if len(X.shape) == 3 :
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]).astype('uint8')
    if len(X.shape) == 4 : 
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]*X.shape[3]).astype('uint8') 
    
    model = ElasticNet()
    model.fit(Xreshape, ytrain)
    return model

def modele_SGD(X,ytrain, normalize=False):
    if len(X.shape) == 2 : 
        Xreshape = X.astype('uint8')
    if len(X.shape) == 3 :
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]).astype('uint8')
    if len(X.shape) == 4 : 
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]*X.shape[3]).astype('uint8') 
    
    model = SGDRegressor(verbose = 1)
    model.fit(Xreshape, ytrain)
    return model

# Fonction de prédiction avec modele

def score_model(modele,X,y):
    if len(X.shape) == 2 : 
        Xreshape = X.astype('uint8')
    if len(X.shape) == 3 :
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]).astype('uint8')
    if len(X.shape) == 4 : 
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]*X.shape[3]).astype('uint8')
    return modele.score(Xreshape,y)

def print_stat(modele,Xtest,ytest, predictions):
    print("MAE : ",mean_absolute_error(ytest,predictions))
    print("MSE : ",mean_squared_error(ytest,predictions))
    print("R^2 : ",score_model(modele,Xtest, ytest))
    
def norm(X):
    if len(X.shape) == 2 : 
        Xreshape = X.astype('uint8')
    if len(X.shape) == 3 :
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]).astype('uint8')
    if len(X.shape) == 4 : 
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]*X.shape[3]).astype('uint8') 
        
    return normalize(Xreshape)
    
    
def prediction_avec_modele(modele, X):
    if len(X.shape) == 2 : 
        Xreshape = X.astype('uint8')
    if len(X.shape) == 3 :
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]).astype('uint8')
    if len(X.shape) == 4 : 
        Xreshape = X.reshape(X.shape[0], X.shape[1]*X.shape[2]*X.shape[3]).astype('uint8') 
    return modele.predict(Xreshape)

# Comparaison des predictions avec la réalité de manière exacte
def comparaison_prediction(predictions, y):
    nb_correctes = 0
    nb_fausses = 0
    nb_images = len(predictions)
    matrice = np.zeros( (8,8), dtype=int) # reel en ligne, prediction en colonne, par tranche de 10 ans
    for i in range(nb_images):
        decenievrai = int(y[i])//10
        deceniepred = int(predictions[i])//10
        #print("vrai")
        #print(decenievrai)
        #print("pred")
        #print(predictions[i])
        #print(int(predictions[i]))
        #print(deceniepred)
        if decenievrai < 8 and decenievrai >= 0 and deceniepred >= 0 and deceniepred < 8 :
            matrice[decenievrai, deceniepred] += 1
            if decenievrai==deceniepred:
                nb_correctes += 1
            else :
                nb_fausses += 1
        else :
            #print("bug")
            nb_fausses += 1
    print(" Matrice de confusion (reel en ligne, prediction en colonne)" )
    print(matrice)
    print("Nombre de bonnes tranches d'âge : " + str(nb_correctes))
    print("Nombre de mauvaises tranches d'âge : " + str(nb_fausses))
    
# Comparaison des predictions avec la réalité à + ou - 10% du vrai age
def comparaison_prediction_approx(predictions, y):
    nb_correctes = 0
    nb_fausses = 0
    nb_images = len(predictions)
    for i in range(nb_images):
        limitebasse = (y[i]*0.9)//1
        limitehaute = (y[i]*1.1)//1
        if (predictions[i]>= limitebasse) and (predictions[i]<= limitehaute) :
            nb_correctes += 1
        else :
            nb_fausses += 1
    print("Précision à 10% : ")
    print("Nombre de réponse valides : " + str(nb_correctes))
    print("Nombre de réponse non valides : " + str(nb_fausses))
    
```

```python
# Normalise les données
scaler = MinMaxScaler()
hog_train_scaled = scaler.fit_transform(hog_features_train, ytrain)
hog_val_scaled = scaler.transform(hog_features_val)
hog_test_scaled = scaler.transform(hog_features_test)
```

Nous allons créer quatre reseaux de perceptron avec Pytorch, chacun avec une structure des couches différentes.

```python
class Net1(nn.Module):
    
    def __init__(self):
        super(Net1,self).__init__()
        self.linear_stack = nn.Sequential(
            nn.Linear(hog_train_scaled.shape[1], 54),
            nn.Linear(54, 7),
            nn.Dropout(p=0.2),
            nn.ReLU(),
        )
        
    def forward(self,x):
        logits = self.linear_stack(x)
        return logits

class Net2(nn.Module):

    def __init__(self):
        super(Net2,self).__init__()
        self.linear_stack = nn.Sequential(
            nn.Linear(hog_train_scaled.shape[1], 32),
            nn.ReLU(),
            nn.Dropout(p=0.25),
            nn.Linear(32, 64),
            nn.ReLU(),
            nn.Dropout(p=0.25),
            nn.Linear(64, 7),
        )
        
    def forward(self,x):
        logits = self.linear_stack(x)
        return logits
    
class Net3(nn.Module):

    def __init__(self):
        super(Net3,self).__init__()
        self.linear_stack = nn.Sequential(
            nn.Linear(hog_train_scaled.shape[1], 32),
            nn.ReLU(),
            nn.BatchNorm1d(32),
            nn.Dropout(p=0.25),
            nn.Linear(32, 64),
            nn.ReLU(),
            nn.BatchNorm1d(64),
            nn.Dropout(p=0.25),
            nn.Linear(64, 128),
            nn.ReLU(),
            nn.BatchNorm1d(128),
            nn.Dropout(p=0.25),
            nn.Linear(128, 7),
        )
        
    def forward(self,x):
        logits = self.linear_stack(x)
        return logits
    
class Net4(nn.Module):

    def __init__(self):
        super(Net4,self).__init__()
        self.linear_stack = nn.Sequential(
            nn.Linear(hog_train_scaled.shape[1], 512),
            nn.ReLU(),
            nn.Dropout(p=0.5),
            nn.Linear(512, 7),
            nn.Sigmoid(),
        )
        
    def forward(self,x):
        logits = self.linear_stack(x)
        return logits
```

```python
def training(n_epochs, data_train, y_train, data_valid,y_valid, model, criterion, optimizer,scheduler=False):
    n_batches_train=(len(data_train)//size_batches) +1
    n_batches_valid=(len(data_valid)//size_batches) +1
    valid_loss_min=100000
    train_losses, valid_losses = [], []
    
    
    for epoch in range(n_epochs):
        train_loss, valid_loss = 0, 0 # monitor losses
      
    
      # train the model
        model.train() # prep model for training
        for i in range(n_batches_train) :
            data = torch.from_numpy(data_train[i*size_batches:(i+1)*size_batches])
            label = torch.from_numpy(y_train[i*size_batches:(i+1)*size_batches])
            
            data = data.to(device=device, dtype=torch.float32)
            label = label.to(device=device, dtype=torch.long)
            
            optimizer.zero_grad() # clear the gradients of all optimized variables
            
            output = model(data) # forward pass: compute predicted outputs by passing inputs to the model
            loss = criterion(output, label) # calculate the loss
            loss.backward() # backward pass: compute gradient of the loss with respect to model parameters
            optimizer.step() # perform a single optimization step (parameter update)
            train_loss += loss.item() * data.size(0) # update running training loss
      
      # validate the model
        model.eval()
        for i in range(n_batches_valid) :
            data = torch.from_numpy(data_valid[i*size_batches:(i+1)*size_batches])
            label = torch.from_numpy(y_valid[i*size_batches:(i+1)*size_batches])
            data = data.to(device=device, dtype=torch.float32)
            label = label.to(device=device, dtype=torch.long)
            
            with torch.no_grad():
                output = model(data)
            loss = criterion(output,label)
            valid_loss += loss.item() * data.size(0)
      
      # calculate average loss over an epoch
        train_loss /= len(data_train)
        valid_loss /= len(data_valid)
        train_losses.append(train_loss)
        valid_losses.append(valid_loss)
        if scheduler :
            lr_scheduler.step(valid_loss)
      
        print('epoch: {} \ttraining Loss: {:.6f} \tvalidation Loss: {:.6f}'.format(epoch+1, train_loss, valid_loss))

      # save model if validation loss has decreased
        if valid_loss <= valid_loss_min:
            print('validation loss decreased ({:.6f} --> {:.6f}).  Saving model ...'.format(
            valid_loss_min,
            valid_loss))
            torch.save(model.state_dict(), 'model.pt')
            valid_loss_min = valid_loss
      
    return train_losses, valid_losses
```

# Optimisation du model3_keras

On utilise la librairie **keras-tuner** afin de créer une série de modèles avec différentes paramètres pour l'optimisation. Dans ce cas, on modifie les paramètres:
- nombre de neurones dans les couches denses.
- taux d'apprentissage
- type d'activation pour la dernière couche, soit ReLU ou Sigmoid.

On utilise également l'optimisation SGD et la perte CategoricalCrossentropy.

```python
import kerastuner as kt

def construire_modele(hyperparams):
    model = keras.Sequential()
    
    hp_units = hyperparams.Int('units', min_value=8, max_value=128, step=8)
    hp_activations = hyperparams.Choice('activation', values=['relu','sigmoid'])
    hp_momentum = hyperparams.Float('momentum', min_value=0.0, max_value=1.0, step=0.2)
    
    model.add(layers.Dense(units=hp_units, activation='relu', input_shape=[hog_train_scaled.shape[1]]))
    model.add(layers.BatchNormalization())
    model.add(layers.Dropout(rate=0.25))
    model.add(layers.Dense(units=(hp_units*2), activation='relu'))
    model.add(layers.BatchNormalization())
    model.add(layers.Dropout(rate=0.25))
    model.add(layers.Dense(units=(hp_units*3), activation='relu'))
    model.add(layers.BatchNormalization())
    model.add(layers.Dropout(rate=0.25))
    model.add(layers.Dense(7, activation=hp_activations))
    
    hp_lr = hyperparams.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])
    
    model.compile(optimizer=keras.optimizers.SGD(learning_rate=hp_lr, momentum=hp_momentum),
                  loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                  metrics=['accuracy'])
    
    return model
```

Maintenant, on effectue la recherche des meilleurs hyperparamètres à l'aide de l'optimisateur de Keras **Hyperband**. On ajoute également un callback afin d'arreter l'entrainement si les valeurs des pertes ne diminuent pas à travers 200 époques et on garde les poids qui ont donné les meilleurs résultats.

```python
tuner = kt.Hyperband(construire_modele,
                     objective='val_accuracy',
                     max_epochs=1000,
                     factor=3,
                     overwrite=True)

early_stopping = keras.callbacks.EarlyStopping(
    min_delta=0.001,
    patience=200,
    restore_best_weights=True,)

reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=20, min_lr=0.001)

tuner.search(hog_train_scaled, ytrain,
             epochs=1000, validation_data=(hog_val_scaled, yval),
             callbacks=[early_stopping, reduce_lr], batch_size=200, verbose=2)
```

```python
# Votre code ici
#Défiition du CNN
class CNN1(nn.Module):  

    def __init__(self):
        super(CNN1,self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, stride=1)#Couche convolutionelle
        self.relu1 = nn.ReLU()
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=1)#Couche convolutionelle
        self.relu2 = nn.ReLU()
        self.maxpl1 = nn.MaxPool2d(kernel_size = (2,2))
        
        # Out = (In - Kernel + 2p)/stride +1 D'ou le 20
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, stride=1)#Couche convolutionelle
        self.relu3 = nn.ReLU()
        self.conv4 = nn.Conv2d(128, 256, kernel_size=3, stride=1)#Couche convolutionelle
        self.relu4 = nn.ReLU()
        self.maxpl2 = nn.MaxPool2d(kernel_size = (2,2))
        
        
        
        #Classifier
        self.fc1 = nn.Linear(256*9*9,128)
        self.relu3 = nn.ReLU()
        self.fc2 = nn.Linear(128,7)

     
    def forward(self,x):
        x = self.conv1(x)
        x = self.relu1(x)
        x = self.conv2(x)
        x = self.relu2(x)
        x = self.maxpl1(x)
        
        x = self.conv3(x)
        x = self.relu3(x)
        x = self.conv4(x)
        x = self.relu4(x)
        x = self.maxpl2(x)
    
        x = x.view( x.size(0), -1)
        x = self.fc1(x)  
        x = self.relu3(x)  
        x = self.fc2(x)   
        return x
```

```python
def print_result_per_batch(model,Xtrain,ytrain,Xval,yval,Xtest,ytest,printed=False):
    model.eval()
    #print(ytrain)
    
    
    #scaler = MinMaxScaler()
    #scaler.fit(Xtrain)
    L=[[Xtrain,ytrain,"train"],[Xval,yval,"val"],[Xtest,ytest,"test"]]
    for datasets in L :
        datas,labels=datasets[0],datasets[1]

        n_batches=(len(datas)//size_batches) +1
        preds = np.zeros( labels.shape[0], dtype=int)    

        for i in range(n_batches) :
            
            
            
            label = torch.from_numpy(labels[i*size_batches:(i+1)*size_batches])
            label = label.to(device=device, dtype=torch.long)  
            data = torch.from_numpy(datas[i*size_batches:(i+1)*size_batches])
            data = data.to(device=device, dtype=torch.float32)
            output=model.forward(data)
            _, batch_preds = torch.max(output, 1)
            preds[i*size_batches:(i+1)*size_batches]=np.array(batch_preds.to(device="cpu"))
            


        nb_correctes = 0
        nb_fausses = 0
        nb_images = len(preds)
        for i in range(nb_images):
            if i%250==0 and printed:
                print(i,"/",nb_images)
            if preds[i]==labels[i]:
                #print()
                nb_correctes += 1
            else :
                nb_fausses += 1
        
        
        print(datasets[2] + " - Taux de classification correctes : "+ str(  (nb_correctes/nb_images)*100 ) +"%")
    print_matrice_confusion(preds,labels)
```

```python
def training_CNN(n_epochs, trainloader, validloader, model, criterion, optimizer,scheduler=False,name_save='model.pt',reg=False):
    #n_batches_train=(len(data_train)//size_batches) +1
    #n_batches_valid=(len(data_valid)//size_batches) +1
    valid_loss_min=100000
    train_losses, valid_losses = [], []
    
    
    for epoch in range(n_epochs):
        train_loss, valid_loss = 0, 0 # monitor losses
      
    
      # train the model
        model.train() # prep model for training
        for (data, label) in trainloader:
            #data = torch.from_numpy(data_train[i*size_batches:(i+1)*size_batches])
            #label = torch.from_numpy(y_train[i*size_batches:(i+1)*size_batches])
            
            data = data.to(device=device, dtype=torch.float32)
            if reg :
                label = label.to(device=device, dtype=torch.float32)
            else :
                label = label.to(device=device, dtype=torch.long)
            
            optimizer.zero_grad() # clear the gradients of all optimized variables
            #print(data.shape)
            #print(data)
            
            output = model(data) # forward pass: compute predicted outputs by passing inputs to the model
            loss = criterion(output, label) # calculate the loss
            loss.backward() # backward pass: compute gradient of the loss with respect to model parameters
            optimizer.step() # perform a single optimization step (parameter update)
            train_loss += loss.item() * data.size(0) # update running training loss
      
      # validate the model
        model.eval()
        for (data, label) in validloader:
            #data = torch.from_numpy(data_valid[i*size_batches:(i+1)*size_batches])
            #label = torch.from_numpy(y_valid[i*size_batches:(i+1)*size_batches])
            data = data.to(device=device, dtype=torch.float32)
            label = label.to(device=device, dtype=torch.long)
            
            with torch.no_grad():
                output = model(data)
            loss = criterion(output,label)
            valid_loss += loss.item() * data.size(0)
      
      # calculate average loss over an epoch
        train_loss /= len(trainloader.sampler)
        valid_loss /= len(validloader.sampler)
        
        train_losses.append(train_loss)
        valid_losses.append(valid_loss)
        if scheduler :
            lr_scheduler.step(valid_loss)
      
        print('epoch: {} \ttraining Loss: {:.6f} \tvalidation Loss: {:.6f}'.format(epoch+1, train_loss, valid_loss))

      # save model if validation loss has decreased
        if valid_loss <= valid_loss_min:
            print('validation loss decreased ({:.6f} --> {:.6f}).  Saving model ...'.format(
            valid_loss_min,
            valid_loss))
            torch.save(model.state_dict(), name_save)
            valid_loss_min = valid_loss
      
    return train_losses, valid_losses 
```

```python
class CustomTensorDataset(Dataset):
  def __init__(self, dataset, transform_list=None):
    [data_X, data_y] = dataset
    X_tensor, y_tensor = torch.tensor(data_X), torch.tensor(data_y)
    #X_tensor, y_tensor = Tensor(data_X), Tensor(data_y)
    tensors = (X_tensor, y_tensor)
    assert all(tensors[0].size(0) == tensor.size(0) for tensor in tensors)
    self.tensors = tensors
    self.transforms = transform_list

  def __getitem__(self, index):
    x = self.tensors[0][index]

    if self.transforms:
      #for transform in self.transforms: 
      #  x = transform(x)
      x = self.transforms(x)

    y = self.tensors[1][index]

    return x, y

  def __len__(self):
    return self.tensors[0].size(0)
```

```python
# Constantes
#heightEN = 224  # Le model efficientNetB0 qui est utilisé nécessite des images de taille 224x224x3
#widthEN = 224

heightEN = 48
widthEN = 48
channels = 3
input_shape = (heightEN,widthEN,channels)
n_classes = 7
```

```python
base_model = keras.applications.EfficientNetB0(
    weights="imagenet",  # Load weights pre-trained on ImageNet.
    input_shape=input_shape,
    include_top=False, # Do not include the ImageNet classifier at the top.
)  

# Freeze the base_model
base_model.trainable = False

# Create new model on top
inputs = keras.Input(shape=input_shape)

# The base model contains batchnorm layers. We want to keep them in inference mode
# when we unfreeze the base model for fine-tuning, so we make sure that the
# base_model is running in inference mode here.
x = base_model(inputs, training=False)
x = keras.layers.GlobalAveragePooling2D()(x)
x = layers.BatchNormalization()(x)
x = keras.layers.Dropout(0.2)(x)  # Regularize with dropout
outputs = layers.Dense(n_classes, activation="softmax", name="pred")(x)
model = keras.Model(inputs, outputs)

model.summary()
```

```python
def plot_hist(hist):
    plt.plot(hist.history["accuracy"])
    plt.plot(hist.history["val_accuracy"])
    plt.title("model accuracy")
    plt.ylabel("accuracy")
    plt.xlabel("epoch")
    plt.legend(["train", "validation"], loc="upper left")
    plt.show()
```

```python
# Normalisation
Xtrain /= 255.
Xval /= 255.
Xtest /= 255.
```

```python
Xtrain_CNNPT = np.reshape(Xtrain , (Xtrain.shape[0],48,48))
Xval_CNNPT = np.reshape(Xval , (Xval.shape[0],48,48))
Xtest_CNNPT = np.reshape(Xtest , (Xtest.shape[0],48,48))

print(Xtrain_CNNPT.shape)

Xtrain_CNNPT = np.repeat(Xtrain_CNNPT[..., np.newaxis], 3, -1)
Xval_CNNPT = np.repeat(Xval_CNNPT[..., np.newaxis], 3, -1)
Xtest_CNNPT = np.repeat(Xtest_CNNPT[..., np.newaxis], 3, -1)

print(Xtrain_CNNPT.shape)
```

```python
ytrainCNNPT = tf.keras.utils.to_categorical(ytrain, 7)
yvalCNNPT = tf.keras.utils.to_categorical(yval, 7)
ytestCNNPT = tf.keras.utils.to_categorical(ytest, 7)
```

```python
# TRAINING
val_data = (Xval_CNNPT, yvalCNNPT)

model.compile(
    
    optimizer=keras.optimizers.Adam(learning_rate = 1e-3, beta_1 = 0.9, beta_2 = 0.999, epsilon = 1e-08, decay = 0.0),
    loss='categorical_crossentropy',
    metrics=['accuracy'],
)

#keras.optimizers.Adam(),
#loss=keras.losses.CategoricalCrossentropy(),
#metrics=[keras.metrics.CategoricalAccuracy()]

epochs = 15
hist = model.fit(Xtrain_CNNPT, ytrainCNNPT, epochs=epochs, validation_data=val_data)
```

```python
plot_hist(hist)
```

```python
# FINE-TUNING

# Unfreeze the base_model. Note that it keeps running in inference mode
# since we passed `training=False` when calling it. This means that
# the batchnorm layers will not update their batch statistics.
# This prevents the batchnorm layers from undoing all the training
# we've done so far.
base_model.trainable = True
model.summary()

model.compile(
    optimizer=keras.optimizers.Adam(1e-5),  # Low learning rate
    loss='categorical_crossentropy',
    metrics=['accuracy'],
)

epochs = 10
#model_history = model.fit(Xtrain_CNNPT, ytrainCNNPT, epochs=epochs, validation_data=val_data)
from keras.callbacks import ModelCheckpoint

checkpointer = ModelCheckpoint(filepath='model.weights.best.hdf5', verbose = 1, save_best_only=True)
model_history = model.fit(Xtrain_CNNPT,
                     ytrainCNNPT,
                     epochs=epochs,
                     validation_data=(Xval_CNNPT, yvalCNNPT),
                     callbacks=[checkpointer])
```

Comme expliqué à l'issue de la première partie, de simples régressions, quelque soit le modèle, sur les données brutes ne permettent pas d'avoir un modèle suffisamment généralisable. En effet, on remarque sur tous les modèles un grossier overfitting, même après cross-validation. Nous avons noté 3 raisons principales à ces mauvais résultat comme mentionné plus haut :
- Les données ne sont pas bien réparties ce qui pousse le modèle à prédire principalement dans les 3 premières tranches d'âge : la régression se fait de manière à minimiser MSE, ou autre fonction loss, sur l'ensemble d'entraînement qui est principalement composé d'image de jeunes personnes ce qui implique donc de minimiser la MSE, ou autre fonction loss, surtout sur les personnes jeunes. Cela force la regression à être réglé de façon à prédire principalement ces personnes là. 
- Traiter directement les données des pixels sans primitives tel hog, lbp ou gabor n'est pas efficace pour une régression dans ce cas d'utilisation. Les valeurs de chaque pixel prise individuellement ne semble pas permettre de prédire l'âge à partir d'un visage efficacement. En effet, le modèle doit principalement étudier, pour cette utilisation, les trais du visages et des rapports concernant les différentes zones du visage qui pourraient permettre de déterminer l'âge correctement. L'intensité de chaque pixel ne semble pas être très efficace ici, surtout pour des photos prises de point de vue pas toujours identiques et comprenant des arrières plans différents.
- Le faible taille du dataset initial n'aide également pas à avoir un modèle très généralisable. Sur 82 personnes, nous avons choisi d'en garder 12 en test et les 70 autres pour notre set d'entraînement. Ceci implique une regression très spécialisée sur ces personnes. Avec un dataset bien plus grand (+ de 1000 personnes), une simple regression comme nous avons utilisé aurait surement été bien plus efficace.

On remarque que les CNN ont un peu mieux performés que le MLP, alors que la recherche d'hyperparamètre était moins poussé pour les CNN. Ceci vient donc bien confirmer l'utilité des CNN pour l'analyse d'images. Cependant on obtient un taux de classification de 55% ce qui reste relativement faible. On peut noter que la data augmentation nous a permis de bien améliorer nos résultats.
On remarque cependant que le transfert learning a moins bien marché, ceci s'explique surtout pas le fait que notre étude n'est pas abouti, et que le modèle utilisé n'était probablement pas adapté. 
Les résultats obtenus avec les réseaux de neurones sont supérieurs aux résultats obtenus lors du tp 1, où nous avions atteints les 48% au maximum.
Afin d'améliorer davantage nos résultats il aurait été intéressant d'essayer de nouvelles architectures de CNN.

On peut aussi noter que les réseaux de neurones ont tendance à être gourmand en ressources, ainsi malgré l'utilisation de gpu, les cnn peuvent prendre jusqu'a 1h d'entraînement. Au vu du faible gain de précision de nos algorithme mais de la perte considérable de temps par rapport à des algorithmes plus classiques, on peut critiquer l'utilisation des réseaux de neurones. Cependant nos résultats pourraient probablement être grandement améliorés avec de meilleures modèles. Parallèement à la complexité temporelle, on peut aussi observer la complexité spatiales de nos réseaux. En effet nos 3 derniers réseaux possèdaient de 3 à 4 millions de paramètres, ce qui est relativement faible par rapport à de gros réseaux tels que VGG, (10 à 20 millions de paramètres ) mais énorme face à des modèles plus classiques de machine learning.
